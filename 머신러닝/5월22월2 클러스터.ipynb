```python
from sklearn.cluster import KMeans
# k네이버스에서 k네이버스는 이웃의 개수
# 마찬가지로 군집의 개수를 말함.

# 사이킷런 계열은 주로 2차원을 다룸. 그러므로 2차원 데이터로 바꿔주기.
```

```python
data_2dim = data.reshape(data.shape[0],data.shape[1]*data.shape[2])
```

```python
km = KMeans(n_clusters=3,random_state=100)
km.fit(data_2dim)
```

```python
km.labels_     # 라벨로 지정함. 사과 파인 바나나가 아니라 0 1 2로 지정하는 것.
```

```python
np.unique(km.labels_, return_counts=True)
 # 0은 111 1은 98 2는 91개  답은 100, 100 , 100인데 머신러닝은 이 정도 수준으로 예측한 것.
```

```python
def showKmeansClustImg(label=0):
  data_len = len(data[km.labels_ == label] )
  rowcol = np.int(np.round(np.sqrt(data_len)))
  for idx,img in enumerate(data[km.labels_ == label]) :
    plt.subplot(rowcol,rowcol,idx+1)
    plt.imshow(img,cmap='gray_r')
    plt.xticks([])
    plt.yticks([])
  plt.show()
```

```python
showKmeansClustImg(1)
```

```python
showKmeansClustImg(2)
```

```python
showKmeansClustImg(0)
```

```python
# 클러스터의 중심

# 클러스터 또한 평균이미지처럼 분류하기 위한 중심이미지를 갖고 있는데
km.cluster_centers_  
# 분류하는 기준이 되는 클러스터의 중심 데이터들.
km.cluster_centers_.shape
# 이걸 이미지로 그려보려면 reshape해서 복구해야함.

center_imgs = km.cluster_centers_.reshape(-1,100,100)
center_imgs
```

```python
center_imgs.shape
```

```python
for idx, img in enumerate(center_imgs):
  plt.subplot(1,3,idx+1)
  plt.imshow(img,cmap='gray_r')
  plt.xticks([])
  plt.yticks([])
plt.show()
```

```python
# 엘 보우 inertia(이너셔) 군집을 결정하는 중심과 각 데이터의 거리의 제곱합. 이너셔가 이걸 결정하는 기준이 되는거.
# 얼마나 가깝게 모여있는지를 나타내는 지표. 그러니깐 작을 수록 응집도가 높은 것임.
# 이 값을 작은 방향으로 학습해서 최적의 군집을 찾는 것. 군집의 갯수는 우리가 정해주어야 함.
# 그래프가 팔 처럼 꺽이는 부분이 최적의 점 이때 값이 변함. 군집의 개수

result = []
for k in range(2,6):
  km = KMeans(n_clusters=k,random_state=100,n_init=10)
  km.fit(data_2dim)
  result.append(km.inertia_)
```

```python
plt.plot(range(2,6),result)
plt.show()
# 꺾이는 폭이 3이 제일 큼.
```
